### Step 2: Use the `str.replace` function to remove the dollar signs ($) from the values in the Close column.

```python
# Use the str.replace function to remove the dollar sign, $
# YOUR CODE HERE

# Thisd function remove_dollar_sign will strip off "$" from the DataFrame colum
# arguments: input_df (DataFrame) , string_out, replacement_string, optional datatype for changing the column data type
def remove_dollar_sign(input_df, string_column_name, string_out, replacement_string, datatype):
    
    # Check if the input arguments
    if (len(input_df) > 0 and (string_out != "") ):
        #input_df.loc[:,string_column_name] = input_df.loc[:,string_column_name].str.replace(string_out, replacement_string)
        input_df.loc[ : ,string_column_name] = input_df.loc[ : ,string_column_name].str.replace(string_out, replacement_string)
            
        # if datatype is provided then change datatype of column to the argumetn of datatype passed    
        if (datatype != "" ) or (datatype != "" ):
            input_df = input_df.loc[:,string_column_name].astype(datatype)
            return input_df
        else:
            input_df = input_df.loc[:,string_column_name]
            return input_df
        #return input_df.loc[:,string_column_name]
    else:
        print(f"ERROR: string for removal can't be missing or null,  {string_out}" )
        print(f"ERROR: input DataFrame can't be empty or null,  {input_df}" )
        
            
            
            
    # Check data to see if the string "$" was removed

bitstamp_clean_df = remove_dollar_sign(bitstamp_clean_df, "Close", "$", "", "float")
bitstamp_clean_df.head()
```

### Step 4: Review the data for duplicated values, and drop them if necessary.

```python
# Review the data for duplicate values, and drop them if necessary
# YOUR CODE HERE
# use duplicated and sum function to check the count of duplicate data


# The following function drop_duplicate_data takes in DataFrame input and strips out duplicate data before returning the DataFram
def drop_duplicate_data(input_df):
    
    # use duplicated and sum function to get check if there are duplicates
    if (input_df.duplicated().sum() == 0):
        # There are no duplicates. Resturn the DataFrame as is
        
        return input_df
    
    elif (input_df.duplicated().sum() > 0):
        # There are duplicate data. Drop duplicate data
        input_df = input_df.drop_duplicates()
        return input_df

        
bitstamp_clean_df = drop_duplicate_data(bitstamp_clean_df)  

# Review or check if there are still duplicates
bitstamp_clean_df.head()
```
